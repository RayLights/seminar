```python
import pystac_client
from odc import stac as odc_stac

catalog_url = 'https://stac.terrabyte.lrz.de/public/api'
collection = 'cop-dem-glo-90'

catalog = pystac_client.Client.open(
    catalog_url
)

grand_canyon = [-112.15,36.05]
search = catalog.search(
    collections = [collection],
    intersects = {"type":"Point","coordinates": grand_canyon},
)
items = list(search.items())
```

```python
print(f"Returned {len(items)} items")
list(items[0].assets)
cube = (odc_stac.load(items).squeeze().drop("time"))
```

```python
import xrspatial 
from datashader.transfer_functions import shade,stack
from datashader.colors import Elevation

hillshade = xrspatial.hillshade(cube.data)
stack(shade(hillshade,cmap["white","gray"]), shade(cube.data,cmap=Elevation,alpha= 128))
```
